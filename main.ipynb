```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/kerala_covid.csv")

# Filter Kerala data
kerala_df = df[df["State"] == "Kerala"].copy()

# Convert date column to datetime
kerala_df["Date"] = pd.to_datetime(kerala_df["Date"])

# Calculate active cases
kerala_df["Active_cases"] = (
    kerala_df["Confirmed"] - kerala_df["Recovered"] - kerala_df["Deceased"]
)


# Select required columns
processed_df = kerala_df[["Date", "Active_cases", "Recovered", "Deceased"]]

# Sort by date
processed_df = processed_df.sort_values(by="Date")

# Save cleaned dataset
processed_df.to_csv("data/kerala_processed_covid.csv", index=False)

# Display first rows
print(processed_df.head())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("data/kerala_population.csv")

# Keep only district and total population columns
df = df[["District", "Population Total"]]

# Convert population from Lakhs to actual numbers
df["Population"] = df["Population Total"] * 100000

# Drop old column
df = df.drop(columns=["Population Total"])

# Rename district column
df = df.rename(columns={"District": "district"})

# Save cleaned dataset
df.to_csv("data/kerala_processed_population.csv", index=False)

# Display first rows
print(df.head())
```

* Hospitals: This column includes the total count of District Hospitals (DH), General Hospitals (GH), Women & Child Hospitals (W&C), and Taluk level hospitals (THQH/TH) within that district under the DHS.

* Exclusions: This count does not include Primary Health Centres (PHCs) or Family Health Centres (FHCs), as those are generally considered clinics rather than full-scale hospitals in the infrastructure summary

* Hospital Beds: These figures represent the sanctioned bed strength in government institutions (DH, GH, THQH, and TH) under the Directorate of Health Services.

* ICU/Ventilators: These are the active functional units reported in the infrastructure census section of the document.

* Source: DHS Kerala, Health at a Glance 2022-23, Tables 3.1 through 3.8.
